```python
import pandas as pd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
heart_data = pd.read_csv('/content/drive/MyDrive/Dataset/heart.csv')
```

```python
heart_data
```

```python
# Data Cleaning
# Check for missing values
```

```python
missing_values = heart_data.isnull().sum()
print("Missing Values:\n", missing_values)
```

```python
# Remove duplicates if any
```

```python
heart_data_cleaned = heart_data.drop_duplicates()
```

```python
# Error Correction
# Identify unrealistic values in certain columns (like 'age', 'chol', 'trestbps')
```

```python
# Check for outliers in 'age' (e.g., age < 18 or age > 100 might be unrealistic)
```

```python
age_outliers = heart_data_cleaned[(heart_data_cleaned['Age'] < 18) | (heart_data_cleaned['Age'] > 100)]
print("\nAge Outliers:\n", age_outliers)
```

```python
heart_data
```

```python
# Check for outliers in 'chol' (e.g., cholesterol < 100 or cholesterol > 600)
```

```python
chol_outliers = heart_data_cleaned[(heart_data_cleaned['Cholesterol'] < 100) | (heart_data_cleaned['Cholesterol'] > 600)]
print("\nCholesterol Outliers:\n", chol_outliers)
```

```python
# Check for outliers in 'trestbps' (e.g., resting blood pressure < 50 or > 250)
```

```python
heart_data
```

```python
trestbps_outliers = heart_data_cleaned[(heart_data_cleaned['RestingBP'] < 50) | (heart_data_cleaned['RestingBP'] > 250)]
print("\nResting Blood Pressure Outliers:\n", trestbps_outliers)
```

```python
# Display basic information about cleaned data
```

```python
print("\nCleaned Dataset Info:")
heart_data_cleaned.info()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=heart_data_cleaned, x='Age', bins=20, kde=True)
plt.title("Age Distribution of Patients")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=heart_data_cleaned, x='HeartDisease', y='Cholesterol')
plt.title("Cholesterol Levels by Heart Disease Status")
plt.xlabel("Heart Disease (0=No, 1=Yes)")
plt.ylabel("Cholesterol Level")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=heart_data_cleaned, x='RestingBP', bins=20, kde=True)
plt.title("Distribution of Resting Blood Pressure")
plt.xlabel("Resting Blood Pressure (mmHg)")
plt.ylabel("Count")
plt.show()
```

```python


plt.figure(figsize=(12, 6))
sns.boxplot(data=heart_data_cleaned, x='ChestPainType', y='MaxHR')
plt.title("Max Heart Rate by Chest Pain Type")
plt.xlabel("Chest Pain Type")
plt.ylabel("Max Heart Rate")
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(heart_data_cleaned[['Age', 'RestingBP', 'Cholesterol', 'MaxHR', 'Oldpeak', 'HeartDisease']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Key Variables")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=heart_data_cleaned, x='ST_Slope', hue='HeartDisease')
plt.title("ST Slope by Heart Disease Status")
plt.xlabel("ST Slope")
plt.ylabel("Count")
plt.legend(title='Heart Disease')
plt.show()
```
