**Assignment 04 :** Implement normal distribution in python and visualize it for Mean =100, standard _deviation =4, dataset size=100000

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
```

```python
mean = 100
std_dev = 4
dataset_size = 100000

data = np.random.normal(mean, std_dev, dataset_size)

plt.figure(figsize=(8, 6))
plt.hist(data, bins=50, density=True, color='skyblue', edgecolor='black')
plt.title('Normal Distribution (Mean=100, Std Dev=4, Dataset Size=100000)')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

```python


mean = 100
std_dev = 4
size = 100000

data = np.random.normal(mean, std_dev, size)

plt.figure(figsize=(10, 6))
sns.histplot(data, bins=50, kde=False, stat='density', color='skyblue', label='Histogram')

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean, std_dev)
plt.plot(x, p, 'r', linewidth=2, label='PDF')

plt.title(f'Normal Distribution (Mean = {mean}, Std Dev = {std_dev}, Size = {size})')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python

mean = 100
std_dev = 4
size = 100000

data = np.random.normal(mean, std_dev, size)

plt.figure(figsize=(12, 7))
sns.histplot(data, bins=50, kde=True, stat='density', color='skyblue', edgecolor='black', alpha=0.7)

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean, std_dev)
plt.plot(x, p, 'r', linewidth=2, label='PDF', zorder=5)

plt.axvline(mean, color='green', linestyle='--', linewidth=2, label='Mean', zorder=3)
plt.axvline(mean + std_dev, color='orange', linestyle='--', linewidth=1, label='1 Std Dev', zorder=3)
plt.axvline(mean - std_dev, color='orange', linestyle='--', linewidth=1, zorder=3)

plt.title('Normal Distribution Visualization', fontsize=16)
plt.xlabel('Value', fontsize=14)
plt.ylabel('Density', fontsize=14)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python

mean = 100
std_dev = 4
size = 100000

data = np.random.normal(mean, std_dev, size)

plt.figure(figsize=(12, 7))

sns.histplot(data, bins=50, stat='density', color='skyblue', edgecolor='black', alpha=0.7, label='Histogram')

x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 100)
p = norm.pdf(x, mean, std_dev)
sns.lineplot(x=x, y=p, color='red', linewidth=2, label='PDF')

plt.axvline(mean, color='green', linestyle='--', linewidth=2, label='Mean', zorder=3)
plt.axvline(mean + std_dev, color='orange', linestyle='--', linewidth=1, label='1 Std Dev', zorder=3)
plt.axvline(mean - std_dev, color='orange', linestyle='--', linewidth=1, zorder=3)

plt.title('Normal Distribution Visualization', fontsize=16)
plt.xlabel('Value', fontsize=14)
plt.ylabel('Density', fontsize=14)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
